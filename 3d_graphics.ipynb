# Imports

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
```

# 3D Scatter Plot

```python
def scatter_3d(x, y, t, xlabel, ylabel, zlabel, title):
  fig = plt.figure(1)
  ax = fig.add_subplot(111, projection='3d')
  ax.plot(xplot, yplot, t ,'r-*')
  ax.set_xlabel(xlabel)
  ax.set_ylabel(ylabel)
  ax.set_zlabel(zlabel)
  ax.set_title(title)


xplot = []
yplot = []
tplot = []

for n in range(100):
  tplot.append(.1*n*math.pi)
  xplot.append(math.cos(tplot[n]))
  yplot.append(math.sin(tplot[n]))

scatter_3d(xplot, yplot, tplot, 'cos((t))', 'sin((t))', 't', 'A Helix')
```

# 3D Surface Plot

```python
def surface3d(minx, maxx, xstri, miny, maxy, ystri, color, xlabel, ylabel, zlabel, title):
  fig = plt.figure(2)
  ax = fig.add_subplot(111, projection='3d')
  x = np.arange(minx, maxx, xstri)
  y = np.arange(miny, maxy, ystri)

  X, Y = np.meshgrid(x, y)
  Z = np.exp(-(X**2 + Y**2)/2.0)

  ax.plot_surface(X,Y,Z,linewidth=0, antialiased=True,color=color)
  ax.set_xlabel(xlabel)
  ax.set_ylabel(ylabel)
  ax.set_zlabel(zlabel)
  ax.set_title(title)

surface3d(-5, 5, 0.1, -5, 5, 0.1, 'green', 'x', 'y', 'z', 'title')
```

# 3D Wire Frame Plot

```python
def wire3d(minx, maxx, xstri, miny, maxy, ystri, rstri, cstri):
  fig = plt.figure(2)
  ax = fig.add_subplot(111, projection='3d')
  x = np.arange(minx, maxx, xstri)
  y = np.arange(miny, maxy, ystri)

  X, Y = np.meshgrid(x, y)
  Z = np.exp(-(X**2 + Y**2)/2.0)

  ax.plot_wireframe(X,Y,Z,rstride=rstri,cstride=cstri)

wire3d(-5, 5, 0.1, -5, 5, 0.1, 10, 10)
```
