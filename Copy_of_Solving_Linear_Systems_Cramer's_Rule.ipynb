# Determinants

In mathematics, the `determinant` is a `scalar` value that is a function of the entries of a square matrix. It allows characterizing some properties of the matrix and the linear map represented by the matrix. The determinant of a product of matrices is the product of their determinants (the preceding property is a corollary of this one). The determinant of a matrix A is denoted det(A), det A, or |A|.

## Cramer's Rule

In linear algebra, Cramer's rule is an explicit formula for the solution of a system of linear equations with as many equations as unknowns, valid whenever the system has a unique solution. It expresses the solution in terms of the determinants of the (square) coefficient matrix and of matrices obtained from it by replacing one column by the column vector of right-sides of the equations.

### To do:

Create a function that returns a tuple containing the solutions to a system of 2 equations with 2 unknowns in Standard form (ax + by = c)

5x + 5y = -20 \\
4x - 4y = 8

Has a solution of (-1, -3)

```python
# create a determinant function that takes a 2x3 matrix (ie. list) as an 
# argument and returns a tuple containing the values of the variables

def solution_2x3(matrix):

  coeff_array = [[matrix[0][0], matrix[0][1]], [matrix[1][0], matrix[1][1]]]
  x_array = [[matrix[0][2], matrix[0][1]], [matrix[1][2], matrix[1][1]]]
  y_array = [[matrix[0][0], matrix[0][2]], [matrix[1][0], matrix[1][2]]]
  
  try:
    x = determinant(x_array) / determinant(coeff_array)
    y = determinant(y_array) / determinant(coeff_array)
    
    return x, y
  except ZeroDivisionError:
    return (None, None)

array = [[5, 5, -20], [4, -4, 8]]
solution_2x3(array)
```

```python
array3x3 = [[3, 5, -2], [6, 1, 4], [-3, 4, 7]]

def determinant3x3(matrix):
  ar1 = [[matrix[1][1], matrix[1][2]], [matrix[2][1], matrix[2][2]]]
  ar2 = [[matrix[1][0], matrix[1][2]], [matrix[2][0], matrix[2][2]]]
  ar3 = [[matrix[1][0], matrix[1][1]], [matrix[2][0], matrix[2][1]]]

  deter1 = determinant(ar1)
  deter2 = determinant(ar2)
  deter3 = determinant(ar3)
  return deter1, deter2, deter3

  

determinant3x3(array3x3)
```
