```python
from matplotlib import pyplot as plt
import numpy as np 
import pandas as pd
```

# Files

```python
pitch_raw = 'https://raw.githubusercontent.com/bobg207/Honors_Data_Analysis/master/2018_MLB_Pitching_Stats_II.txt'
hit_raw = 'https://raw.githubusercontent.com/bobg207/Honors_Data_Analysis/master/2018_MLB_Hitting_Stats.txt'
pitch_data = pd.read_csv(pitch_raw, delim_whitespace=True)
hit_data = pd.read_csv(hit_raw, delim_whitespace=True)

hit_data
```

# Moneyball

```python
def correlation(x, y):
  x_sum = 0
  x_squ = 0
  y_sum = 0
  y_squ = 0
  xy_sum = 0

  for n in range(len(x)):
    xy_sum += x[n] * y[n]
    x_sum += x[n]
    x_squ += x[n] * x[n]
    y_sum += y[n]
    y_squ += y[n] * y[n]

  x_y = x_sum * y_sum
  numerator = len(x) * xy_sum - x_y

  subtract_x = (len(x) * x_squ - x_sum**2)**0.5
  subtract_y = (len(y) * y_squ - y_sum**2)**0.5
  denomonator = subtract_x * subtract_y
  corr_coeff = numerator / denomonator
  return corr_coeff

hit_columns = ['WPCT', 'OBP', 'SLG', 'AVG', 'RBI']
pitch_columns = ['WPCT', 'ERA', 'SO', 'ER', 'HR']
hit_corrs = []
pitch_corrs = []
for index in range(1, len(hit_columns)):
  hit_corr = correlation(hit_data[hit_columns[0]], hit_data[hit_columns[index]])
  hit_corrs.append(hit_corr)
  pitch_corr = correlation(pitch_data[pitch_columns[0]], pitch_data[pitch_columns[index]])
  pitch_corrs.append(pitch_corr)

print(hit_corrs)
print(pitch_corrs)
```
